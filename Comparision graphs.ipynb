```python
import matplotlib.pyplot as plt

import numpy as np

x = np.arange(-3*np.pi,3*np.pi,0.01)
import matplotlib.pyplot as plt

y = np.sin(3*x)/x
y2 = np.sin(2*x)/x
y3 = np.sin(x)/x
plt.plot(x,y,color = 'b')
plt.plot(x,y2, color ='r')
plt.plot(x,y3, color = 'g')
plt.xticks = ([-2*np.pi,-1*np.pi,0,1*np.pi,2*np.pi])
[r'$-2\pi$',r'$-1\pi$',r'$0$',r'$+1\pi$',r'$-2\pi$']
plt.yticks=([-1,0,1,2,3])
[r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$']
#plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}=1$,xy=[0,1]')
```
